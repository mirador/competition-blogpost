This notebook calculates the median salary for baseball players born as a functin of the month of birth

```python
import pandas as pd
import calendar
%matplotlib inline
```

```python
alldata = pd.read_csv("birthMonth-Salary.tsv", delimiter="\t", na_values="\\N")
alldata_nonus = alldata[alldata["birthCountry"] != "USA"].drop("birthCountry", 1)
alldata = alldata.drop("birthCountry", 1)
```

```python
alldata.dtypes
```

```python
players = alldata.groupby(["playerID"]).max()
yob_mask = (1931 <= players["birthYear"]) & (players["birthYear"] <= 1993)
players = players[yob_mask]
salary_median = players.groupby(["birthMonth"]).median()["salary"]
salary_quant25 = players.groupby(["birthMonth"]).quantile(0.25)["salary"]
salary_quant75 = players.groupby(["birthMonth"]).quantile(0.75)["salary"]
counts = players.groupby(["birthMonth"]).count()["birthMonth"]
```

```python
#salary_quant25.plot()
#salary_quant75.plot()
salary_median.plot()
```

```python
counts.plot()
```

```python
players_nonus = alldata_nonus.groupby(["playerID"]).max()
counts_nonus = players_nonus.groupby(["birthMonth"]).count()["birthMonth"]
counts_nonus.plot()
```

```python
final_data = pd.DataFrame(data = zip(calendar.month_name[1:], salary_median, salary_quant25, salary_quant75, counts, counts_nonus), columns=["Month", "Median", "25% quartile", "75% quartile", "Count", "Count non-US"])
final_data.to_csv("salary.csv", index=False)
final_data
```
