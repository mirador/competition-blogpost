This notebook generates counts for the Exercise-Health correlation in the BRFSS dataset

```python
import pandas as pd
%matplotlib inline
```

```python
alldata = pd.read_csv("Exercise-Health.tsv", delimiter="\t", na_values="\\N")
```

```python
alldata.dtypes
```

```python
# SEX - 1:Male, 2:Female

age_mask = (18 <= alldata["AGE"]) & (alldata["AGE"] <= 49)

act_mask = alldata["QLACTLM2"] == 2 # Activity Limitation Due to Health Problems - 1:yes, 2:no

# Income level:
# 1:Less than $10,000
# 2:Less than $15,000
# 3:Less than $20,000
# 4:Less than $25,000
# 5:Less than $35,000
# 6:Less than $50,000
# 7:Less than $75,000
# 8:$75,000 or more
in_mask = alldata["INCOME2"] <= 8

ex_mask = alldata["EXERANY2"] <= 2  # Exercise in past 30 days - 1:yes, 2:no
gh_mask = alldata["GENHLTH"] <= 5   # General health - 1:Excellent, 2:Very good, 3:Good, 4:Fair, 5:Poor

respondents = alldata[age_mask & act_mask & in_mask & ex_mask & gh_mask][["SEX", "INCOME2", "EXERANY2", "GENHLTH"]]
respondents
```

```python
sex_masks = {"Male":respondents["SEX"] == 1, "Female":respondents["SEX"] == 2}
inc_masks = {"15k":(respondents["INCOME2"] == 1) | (respondents["INCOME2"] == 2),
             "25k":(respondents["INCOME2"] == 3) | (respondents["INCOME2"] == 4),
             "50k":(respondents["INCOME2"] == 5) | (respondents["INCOME2"] == 6),
             "50k+":(respondents["INCOME2"] == 7) | (respondents["INCOME2"] == 8)}

for sex in sex_masks:
    for inc in inc_masks:
        print "Saving",sex,inc
        subset = respondents[sex_masks[sex] & inc_masks[inc]]
        grouped = subset.groupby(["EXERANY2", "GENHLTH"])
        # Name the counts, and flatten the hierarchical indexing of the group data frame
        # so it can be saved as a regular csv file
        counts = pd.DataFrame({"COUNT" : grouped.size()}).reset_index()
        counts.to_csv("health-" + sex + "-" + inc + ".csv", index=False)
```
