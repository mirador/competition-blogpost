This notebook saves the table with the World Bank data on Research expenditure and number of researchers for countries in Europe & Central Asia, North America, and East Asia & Pacific for the 2000-2013 period.

```python
import pandas as pd
%matplotlib inline
```

```python
alldata = pd.read_csv("Expenditure-Researchers.tsv", delimiter="\t", na_values="\\N")
alldata["RATIO"] = alldata["SP.POP.SCIE.RD.P6"] / alldata["GB.XPD.RSDV.GD.ZS"]
alldata.dtypes
```

```python
year_mask = (2000 <= alldata["YEAR"]) & (alldata["YEAR"] <= 20013)
teach_mask = (50 <= alldata["SE.SEC.TCHR.FE.ZS"]) & (alldata["SE.SEC.TCHR.FE.ZS"] <= 100)
#region_mask = (alldata["REGION"] == "Europe & Central Asia") | (alldata["REGION"] == "North America") | (alldata["REGION"] == "East Asia & Pacific")
region_mask = (alldata["REGION"] == "Europe & Central Asia") | (alldata["REGION"] == "North America")
#countries = alldata[year_mask & region_mask & teach_mask]
countries = alldata[year_mask & region_mask]
#countries = alldata[year_mask & teach_mask]
#countries = alldata[year_mask]
countries
```

```python
countries.plot(kind="scatter", x="GB.XPD.RSDV.GD.ZS", y="SP.POP.SCIE.RD.P6")
```

```python
#var = "SP.POP.SCIE.RD.P6" # Researchers
var = "GB.XPD.RSDV.GD.ZS" # R&D Expenditure

threshold = countries.median()[var]
print threshold

subset = countries[threshold < countries[var]]
q75 = subset.quantile(0.75)["RATIO"]
subset[q75 <= subset["RATIO"]]
print q75
grouped = subset[q75 <= subset["RATIO"]].groupby(["NAME"])
print grouped.size()
```
